```python
# Regions
allowed_regions = ["north", "south", "east", "west"]

# define the ideal custmoer
def ideal_customer():
    while True:
        user_input = input("Enter ideal customer, (name age spend region): ")
        parts = user_input.split()
        if len(parts) != 4:
            print("\033[91mError:\n please enter exactly 4 values.\033[0m")
            continue

        # naming values
        name, age_string, spend_string, region = parts

        # capitalize name
        name = name.capitalize()


        # check age
        try:
          age = int(age_string)
          if age < 18 or age > 100:
            print("\033[91mError:\n Age must be 18–100.\033[0m")
            continue
        except ValueError:
          print("\033[91mError:\n Age must be a number.\033[0m")
          continue

        # check spend
        try:
          spend = float(spend_string)
          if spend <= 0:
            print("\033[91mError:\n Spend must be a positive number.\033[0m")
            continue
        except ValueError:
            print("\033[91mError:\n Spend must be a number.\033[0m")
            continue

        # check region
        # convert to lowercase
        region_input = region.lower()
        if region_input not in allowed_regions:
            print(f"\033[91mError:\n Region must be one of the allowed regions:{allowed_regions}.\033[0m")
            continue
        # capitalize to retunr in proper format
        region = region_input.capitalize()


        return {"name": name, "age": age, "spend": spend, "region": region}

# collect up to 16 other customers
def other_customers():
    customers = []
    while len(customers) < 16:
        user_input = input(f"Enter customer {len(customers)+1} (or type done to finish):")
        # 'done'
        if user_input.lower() == 'done':
          if len(customers) == 0:
            print("\033[91mError:\n Enter at least one customer.\033[0m")
            continue
          else:
            break

        parts = user_input.split()
        if len(parts) != 4:
            print("\033[91mError: Please enter exactly 4 values.\033[0m")
            continue

        name, age_string, spend_string, region = parts

        # capitalize name
        name = name.capitalize()

        # check age
        try:
            age = int(age_string)
            if age < 18 or age > 100:
                print("\033[91mError: Age must be 18–100.\033[0m")
                continue
        except ValueError:
            print("\033[91mError: Age must be a number.\033[0m")
            continue

        # check spend
        try:
            spend = float(spend_string)
            if spend <= 0:
                print("\033[91mError:\n Spend must be a positive number.\033[0m")
                continue
        except ValueError:
            print("\033[91mError:\n Spend must be a number.\033[0m")
            continue

        # check region
        region_input = region.lower()
        if region_input not in allowed_regions:
            print(f"\033[91mError:\n Region must be one of the allowed regions:{allowed_regions}.\033[0m")
            continue
        region = region_input.capitalize()

        # compile valid customer entries in a list
        customers.append({"name":name, "age":age, "spend":spend, "region":region})
        print(f"Customer added: {name} {age} {spend} {region}")

    return customers


# checking if customers match, matches if:
    # age diference is <= 5 years, spend is within +- 20% of the ideal, and in the same region
def customer_match(ideal, customer):
  # check the age is +-5
  age_match = (customer["age"] >= ideal["age"] - 5) and (customer["age"] <= ideal["age"] + 5)
  # check the spend is within 20%
  lower_spend = ideal["spend"] * 0.8
  upper_spend = ideal["spend"] * 1.2
  spend_match = (lower_spend <= customer["spend"] <= upper_spend)
  # check the region is the same
  region_match = (customer["region"] == ideal["region"])
  return age_match and spend_match and region_match


ideal = ideal_customer()
print("\nIdeal customer collected:")
print(ideal)

customers_list = other_customers()
print("\nAll entered customers:")

for customer in customers_list:
    print(customer)
print("\nResults:")
for customer in customers_list:
    if customer_match(ideal, customer):
        print(f"\033[32m{customer['name']} matches the ideal customer.\033[0m")
    else:
        print(f"{customer['name']} does NOT match the ideal customer.")
```
